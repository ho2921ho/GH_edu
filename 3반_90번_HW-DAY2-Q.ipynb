## 함수 bestBanc(comp)를 다음과 같이 만들어보세요.
### 한 보험사의 가장 이율이 높은 보험상품을 찾아주는 함수입니다.
1. 매개변수 comp는 가입할 보험사의 이름입니다.
2. 입력받은 보험사의 보험 중 이율이 가장 높은 것을 찾아 보험의 이름과 이율을 return 해주세요.

```python
import numpy as np
banca = np.array(banca)

def bestBanc(comp):
    candidate = banca[:,1:][banca[:,0] == comp]
    best = candidate[candidate[:,1] == max(candidate[:,1])]
    return (float(best[0][1]),best[0][0])
```

```python
#여기에 bestBanc(comp) 함수를 정의하세요.
print(bestBanc('농협손보'))
print(bestBanc('삼성생명'))
```

## 함수 predGain(year, rate)를 다음과 같이 만들어 보세요.
### 저축기간과 이율을 사용해 수익율을 예측하는 함수입니다.
1. 매개변수 year는 저축년수, rate는 이율입니다. 이율은 <b>연리 복리</b>로 계산해주세요.
2. 입력받은 year 만큼 년수가 지나면 얻을 수 있는 이득을 percentage 값으로 return 해주세요.

```python
year = 5
r = 2.57
def predGain(year,rate):
    total_rate = (1+rate/100)**(year)
    print('{}%'.format((total_rate - 1)*100 ))
    
```

$$ 연복리 =  원금*(1+r)^n $$
$$ r = 수익률, n = 납입연수 $$
