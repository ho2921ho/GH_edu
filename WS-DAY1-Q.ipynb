### Q5
- 다음과 같은 규칙을 가진 숫자의 합을 for문을 사용해서 구현하세요.
   - 1 + (1 + 3) + (1 + 3 + 5) + (1 + 3 + 5 + 7) + ... + (1 + 3 + ... + 19)
