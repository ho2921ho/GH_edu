# 1. 데이터 불러오기

```python
# 상반기 주유소 판매 데이터 불러오기
first = pd.read_csv('./data/2019년 상반기 주유소 판매가격.csv', encoding = 'ms949')
first
```

```python
second = pd.read_csv('./data/2019년 하반기 주유소 판매가격.csv', encoding = 'ms949')
second
```

```python
# 상반기 판매가격 데이터 프레임 결측치 조회
first.isnull().sum()
```

```python
# 하반기 판매가격 데이터 프레임 결측치 조회
second.isnull().sum()
```

# 2. 데이터 결합

```python
# 상반기와 하반기 데이터를 상하로 결합
df = pd.concat([first,second], axis = 0)
df
```

# 3. 데이터 가공 및 분석

```python
# 데이터 프레임 정보 조회
df.info()
```

```python
# 상표 컬럼의 고유값 조회
set(df['상표'])
```

```python
# 지역 컬럼의 고유값 조회
set(df['지역'])
```

```python
# 지역 컬럼의 값을 이용하여 시와 구를 분리
df['시'] = [x[0:2] for x in df['지역']]
df['구'] = [x[3:] for x in df['지역']]
df
```

```python
# 시 컬럼의 서울을 서울특별시로 변환
df['시'] = [x + '특별시' for x in df['시']]
df
```

```python
# 기간 컬럼을 datetime 형식으로 변경
df['기간'] = pd.to_datetime(df['기간'], format = '%Y%m%d')
df
```

```python
# 기간 컬럼에서 dt 타입을 이용하여 년, 월, 일, 요일 컬럼 생성
df['월'] = [x.month for x in df['기간']]
df['년'] = [x.year for x in df['기간']]
df['일'] = [x.day for x in df['기간']]
df['요일'] = [x.weekday() for x in df['기간']]
```

```python
weekday_dict = {0:'Monday',1:'Tuesday',2:'Wednesday', 3: 'Thursday',4:'Friday',5:'Saturday',6:'Sunday'}
```

```python
df['요일'] = [weekday_dict[x] for x in df['요일']]
```
